```python
# Question 1: Take 5 numbers from the user and print how many of them are greater than 100

numbers = [50, 120, 200, 30, 150]

count = 0
for n in numbers:
    if n > 100:
        count += 1

print("Numbers greater than 100:", count)
```

```python
# Question 2: Take a list of numbers and print only the numbers that are less than 0

numbers = [10, -5, 7, -20, 3]

for n in numbers:
    if n < 0:
        print(n)
```

```python
# Question 3: Take a string and print only the uppercase letters

text = "HeLLo WoRLd"

for ch in text:
    if ch.isupper():
        print(ch)
```

```python
# Question 4: Take a list of numbers and print the square of each number

numbers = [2, 4, 6, 8]

for n in numbers:
    print(n**2)
```

```python
# Question 5: Take a list of numbers and check whether all numbers are positive

numbers = [5, 8, 10, 3]

all_positive = True

for n in numbers:
    if n < 0:
        all_positive = False
        break

print("All numbers positive:", all_positive)
```

```python
# Question 8: Take a list of numbers and count how many are even and greater than 50

numbers = [20, 60, 55, 80, 33, 100]

count = 0

for n in numbers:
    if n % 2 == 0 and n > 50:
        count += 1

print("Count:", count)
```

```python
# Question 9: Take a list of numbers and find the second largest number

numbers = [10, 50, 30, 80, 60]

numbers.sort()

print("Second largest:", numbers[-2])
```

```python
# Question 10: Take a list of numbers and replace all negative numbers with 0

numbers = [5, -3, 10, -7, 8]

for i in range(len(numbers)):
    if numbers[i] < 0:
        numbers[i] = 0

print(numbers)
```

```python
# Question 11: Take a string and check whether it contains both letters and digits

text = "abc123"

has_letter = False
has_digit = False

for ch in text:
    if ch.isalpha():
        has_letter = True
    if ch.isdigit():
        has_digit = True

if has_letter and has_digit:
    print("Contains both letters and digits")
else:
    print("Does not contain both")
```

```python
# Question 12: Take a list of numbers and print the difference between maximum and minimum

numbers = [5, 10, 20, 3, 15]

difference = max(numbers) - min(numbers)

print("Difference:", difference)
```

```python
# Question 13: Take a list of ages and print how many people are eligible to vote (age ≥ 18)

ages = [15, 22, 30, 17, 19]

count = 0

for age in ages:
    if age >= 18:
        count += 1

print("Eligible voters:", count)
```

```python
# Question 15: Create a menu-driven program using while loop

numbers = []

while True:

    print("\n1. Add number")
    print("2. Remove number")
    print("3. Display list")
    print("4. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        num = int(input("Enter number to add: "))
        numbers.append(num)

    elif choice == 2:
        num = int(input("Enter number to remove: "))
        if num in numbers:
            numbers.remove(num)
        else:
            print("Number not found")

    elif choice == 3:
        print("List:", numbers)

    elif choice == 4:
        print("Program ended")
        break

    else:
        print("Invalid choice")
```
